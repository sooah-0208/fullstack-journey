```python
from pyspark.sql import SparkSession
```

```python
spark = SparkSession.builder.appName("PySpark WordCount Test").master("spark://master:7077").getOrCreate()
```

```python
print("Spark Version: ", spark.version)
```

```python
data = ["hello spark", "hello docker", "hello worker"]
print(data, type(data))
```

```python
rdd = spark.sparkContext.parallelize(data)
print(rdd, type(rdd))
```

```python
counts = rdd.flatMap(lambda x: x.split(" ")).map(lambda word: (word, 1)).reduceByKey(lambda a,b: a+b)
print(counts, type(counts))
```

```python
for word, count in counts.collect():
    print(f"{word}: {count}")
```

```python
print(counts.collect())
```

```python
print( type(rdd) )
```

```python
counts = rdd.map(lambda x: x.lower()).flatMap(lambda x: x.split(" ")).map(lambda word: (word, 1)).reduceByKey(lambda a, b: a + b)
```
