```python
import torch
import numpy as np
print(torch.__version__)
```

```python
arr = np.array([1, 3, 5, 7, 9])
print(arr)
print(arr.dtype)
print(type(arr))
```

```python
t1 = torch.from_numpy(arr)
print(t1)
print(t1.dtype)
print(t1.type())
print(type(t1))
```

```python
t2 = torch.as_tensor(arr)
print(t2)
print(t2.dtype)
print(t2.type())
print(type(t2))
```

```python
arr[0] = 999
```

```python
print(f't1: {t1}')
print(f't2: {t2}')
```
